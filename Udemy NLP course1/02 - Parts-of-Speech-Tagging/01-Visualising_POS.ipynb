```python
import spacy 
nlp = spacy.load("en_core_web_sm")

from spacy import displacy
```

```python
doc= nlp("The quick brown fox jumped over the lazy dog's back.")
```

```python
# visulaizing the dependency parser
displacy.render(doc,style = "dep")
```

```python
# visulaizing the dependency parser
displacy.serve(doc,style="dep", options={"compact" : True})
```

```python
for token in doc:
    print(f"{token.text:<{10}} {token.pos_:>{7}} {token.dep_:^{7}} {spacy.explain(token.dep_)}")
```

```python
doc = nlp("When Sebastian Thrun started working on self-driving cars at Google in 2007, few people outside of the company took him seriously.")
```

```python
# visualising the entities
displacy.render(doc, style = 'ent')
```

```python
displacy.serve(doc,style = "ent")
```
